### Funo Linear

Uma funo linear  uma funo matemtica que pode ser expressa na forma $f(x) = wx + b$, onde $x$ representa a varivel independente, $w$  a inclinao da reta (tambm conhecida como coeficiente angular), e $b$  o intercepto $y$, ou seja, o ponto em que a reta cruza o eixo $y$ no plano cartesiano. A principal caracterstica de uma funo linear  que ela cria uma linha reta quando gerado um grfico.

Aqui esto algumas caractersticas importantes das funes lineares:

- **Inclinao ou coeficiente angular ($w$):** Determina a inclinao da linha. Se $w$ for positivo, a linha inclina-se para cima  medida que se move da esquerda para a direita. Se $w$ for negativo, a linha inclina-se para baixo.
- **Intercepto y ($b$):**  o ponto onde a linha cruza o eixo y.

### Aprendizado

image.png

**Processo de aprendizado:**

1. **Propagao para frente (Forward Propagation) **: Onde as entradas so passadas atravs da rede, atravs de suas camadas ocultas at a camada de sada para obter a sada prevista.

2. **Clculo do Erro **: A sada prevista  comparada com a sada real e calcula-se a diferena ou erro, geralmente utilizando uma funo de perda como a entropia cruzada ou o erro quadrtico mdio.

3. **Backpropagation **: Neste passo, o erro  propagado de volta pela rede, comeando da sada para as camadas anteriores. O gradiente da funo de perda  calculado em relao a cada peso, o que indica como os pesos devem ser ajustados para reduzir o erro.

4. **Atualizao dos Pesos **: Com o gradiente calculado, os pesos so ajustados.

```python
import torch

w = torch.tensor(0.3, requires_grad=True)
b = torch.tensor(-0.01, requires_grad=True)
x = torch.arange(-10,10)
y = 3.2 *  torch.arange(-10,10) + 3
```

```python
def perceptron(w,x,b):
  return w * x + b
```

```python
w_values, b_values = [], []
for epoch in range(200):

  #Etapa do feed Forward
  w_values.append(w.item())
  b_values.append(b.item())
  output = perceptron(w,x,b)

  #Calcular o erro
  squared_error = (output - y)**2
  error = torch.sum(squared_error)

  print(f"Epoch: {epoch}, Error: {error.item()}")

  #Calcular a derivada parcial
  error.backward()

  # Ajusta os valores de w e b
  with torch.no_grad():
    w -= 0.001 * w.grad
    b -= 0.001 * b.grad

  w.grad.zero_()
  b.grad.zero_()
```

```python
output = perceptron(w,x,b)
output, y
```

### funes

Na matemtica, uma funo  um conceito fundamental que descreve uma relao especial entre conjuntos de nmeros ou mais geralmente, entre elementos de conjuntos quaisquer. Uma funo associa dois conjuntos de elementos de tal forma que a cada elemento do primeiro conjunto, chamado de domnio, corresponde exatamente um elemento no segundo conjunto, chamado de contradomnio ou imagem.

### Caractersticas Principais de uma Funo:

1. **Domnio**: O conjunto de todas as entradas possveis para a funo. Por exemplo, o conjunto de todos os nmeros reais, todos os nmeros inteiros, ou qualquer outro conjunto definido.

2. **Contradomnio**: O conjunto de todas as sadas possveis que a funo pode produzir. Nem todas as sadas possveis precisam ser efetivamente atingidas pela funo; as que so atingidas so chamadas de imagem.

3. **Regra de Correspondncia**: Uma "regra" que define como cada entrada  associada a uma sada. Essa regra pode ser uma expresso matemtica, uma equao, ou uma descrio verbal que pode ser sistematicamente aplicada para determinar a sada a partir de uma entrada.

#### Tipos de Funes:

- **Funes Lineares**: Funes que formam uma linha reta quando graficadas, como $ f(x) = 2x + 3 $.
- **Funes Polinomiais**: Incluem termos de potncias de $ x $, como $ f(x) = x^2 - 4x + 4 $.
- **Funes Trigonomtricas**: Envolve funes seno, cosseno, tangente, etc., como $ f(x) = sin(x)$.
- **Funes Exponenciais e Logartmicas**: Involvem exponenciao e logaritmos, como $ f(x) = e^x $ ou $ f(x) = \log(x) $.

#### Importncia das Funes:

Funes so fundamentais em todas as reas da matemtica e suas aplicaes, incluindo fsica, engenharia, economia, biologia, e muitas outras cincias. Elas ajudam a modelar relaes e processos do mundo real, permitindo previses e compreenses profundas de fenmenos naturais e humanos.

Em resumo, uma funo  uma maneira de expressar uma dependncia entre quantidades, onde a variao de uma quantidade determina a variao de outra.

### Derivada

A derivada de uma funo em um determinado ponto  a inclinao da reta tangente  funo nesse ponto. Em termos mais formais, a derivada de uma funo $ f(x) $ em relao a $ x $  definida como o limite da razo entre a mudana de $ f(x) $ e a mudana de $ x $ quando $ x $ tende a zero. A expresso matemtica para a derivada :

$ f'(x) = \lim_{\Delta x \to 0} \frac{f(x + \Delta x) - f(x)}{\Delta x} $

Esta expresso calcula a taxa instantnea de mudana da funo $ f(x) $ em relao a $ x $. Em outras palavras, nos d a inclinao da reta tangente ao grfico da funo no ponto $ x $. A reta tangente  a reta que "apenas toca" o grfico da funo em um nico ponto, e a inclinao dessa reta tangente  um indicador da rapidez com que a funo est aumentando ou diminuindo nesse ponto.

### **Regra da cadeia**

A regra da cadeia  uma tcnica fundamental no clculo diferencial que  utilizada para derivar funes compostas. Essencialmente, ela permite calcular a derivada de uma funo que  o resultado da composio de duas ou mais funes.

#### Forma Normal (Propagao Direta da Derivada) 1 variavel

Neste mtodo, as derivadas so calculadas diretamente de acordo com a composio das funes.

Vamos usar as funes $ b(x) $ e $ n(x) $ e aplicar a regra da cadeia para calcular a derivada de uma funo composta utilizando esses nomes de funo.

**Definio das Funes**

Suponhamos que:

- $ b(x) $ seja uma funo de $ x $.
- $ n(x) $ seja outra funo de $ x $.
- Queremos derivar uma funo composta $ f(x) = b(n(x)) $.

**Aplicao da Regra da Cadeia**

1. **Identificar as funes:**
   - $ n(x) $  a funo interna.
   - $ b(u) $  a funo externa, onde $ u = n(x) $.

2. **Derivar cada funo:**
   - Derive $ n(x) $ em relao a $ x $: $ n'(x) $.
   - Derive $ b(u) $ em relao a $ u $: $ b'(u) $.

3. **Aplicar a regra da cadeia:**
   - A derivada da funo composta $ f(x) $  dada por:
$
     f'(x) = b'(n(x)) \cdot n'(x)
$
     

**Exemplo Prtico**

Suponha que:

- $ n(x) = x^2 $ (a funo interna).
- $ b(u) = \sin(u) $ (a funo externa).

Ento:

1. **Funo Composta:** $ f(x) = \sin(x^2) $.
2. **Derivada da funo interna $ n(x) $:**
   - $ n'(x) = 2x $.
3. **Derivada da funo externa $ b(u) $ quando $ u = x^2 $:**
   - $ b'(u) = \cos(u) $.
4. **Aplicando a regra da cadeia:**
   - $ f'(x) = \cos(x^2) \cdot 2x $.

### Descida do Gradiente

A derivada parcial de uma funo em relao a uma varivel indica a taxa de variao de $ f $ em relao quela varivel, mantendo as outras constantes. O sinal da derivada parcial pode te dar informaes importantes:

- Se a derivada parcial  positiva ($ + $), isso indica que um pequeno aumento na varivel leva a um aumento no valor de $ f $. Ou seja, $ f $ cresce  medida que a varivel aumenta.
- Se a derivada parcial  negativa ($ - $), isso significa que um pequeno aumento na varivel leva a uma diminuio no valor de $ f $. Portanto, $ f $ diminui  medida que a varivel aumenta.

#### O que  o Gradiente?
O gradiente de uma funo $ f(x_1, x_2, \ldots, x_n) $ em um ponto especfico  um vetor que contm todas as derivadas parciais da funo naquele ponto. Formalmente,  definido como:
$$
\nabla f = \left( \frac{\partial f}{\partial x_1}, \frac{\partial f}{\partial x_2}, \ldots, \frac{\partial f}{\partial x_n} \right)
$$
Este vetor no apenas indica a direo de maior aumento, mas sua magnitude tambm fornece uma noo da taxa de aumento da funo nessa direo.

Geometricamente, voc pode imaginar o gradiente como uma seta que aponta para cima a partir de um ponto em uma superfcie montanhosa. Se voc seguir na direo dessa seta, estar subindo a montanha pelo caminho mais ngreme.

Quando todas essas mudanas so combinadas (como no vetor gradiente), elas apontam para a direo onde a funo cresce mais rapidamente se todas as variveis forem alteradas proporcionalmente ao valor de suas respectivas derivadas parciais.

### Descida do Gradiente

Na tcnica de otimizao conhecida como "descida do gradiente", o objetivo  encontrar o minmo de uma funo. Para fazer isso, utilizamos o fato de que o gradiente aponta para o maior aumento e, portanto, o oposto do gradiente (ou gradiente negativo) apontar para a maior diminuio. Isso significa que, se movermos na direo oposta ao gradiente, estamos movendo na direo que reduz o valor da funo mais rapidamente.


```python
import numpy as np
import matplotlib.pyplot as plt

# Funo e gradiente
def f(x):
    return x**2 - 4*x + 4

def grad_f(x):
    return 2*x - 4

# Descida do gradiente
def gradient_descent(initial_x, learning_rate, n_steps):
    x = initial_x
    trajectory = [x]
    for _ in range(n_steps):
        gradient = grad_f(x)
        x = x - learning_rate * gradient
        trajectory.append(x)
    return trajectory

# Gerar dados
x_values = np.linspace(-1, 5, 400)
y_values = f(x_values)
trajectory = gradient_descent(0.5, 0.1, 15)

# Gerar o grfico
plt.figure(figsize=(8, 5))
plt.plot(x_values, y_values, label='f(x) = x - 4x + 4')
plt.scatter(trajectory, f(np.array(trajectory)), color='red', s=50, label='Descida do Gradiente')
plt.plot(trajectory, f(np.array(trajectory)), color='red', linestyle='--', alpha=0.5)
plt.title('Exemplo de Descida do Gradiente')
plt.xlabel('x')
plt.ylabel('f(x)')
plt.legend()
plt.grid(True)
plt.show()
```

## Teorema da Aproximao Universal

**O que  o Teorema da Aproximao Universal?**

O Teorema da Aproximao Universal afirma que uma rede neural com uma nica camada oculta pode aproximar qualquer funo contnua em espaos compactos. Isso significa que, com a arquitetura adequada, uma rede neural pode teoricamente modelar qualquer relao matemtica, no importando sua complexidade.

**Aplicao Prtica: Modelando o Lucro de uma Empresa**

Suponha que desejamos entender ou prever o lucro de uma empresa com base em fatores como vendas, custos e impostos. Utilizando o teorema da aproximao universal, podemos configurar uma rede neural para aprender a funo que liga esses inputs ao lucro.

1. **Coleta de Dados**: Primeiramente, coletamos dados histricos que incluem vendas, custos, e outros fatores relevantes junto com os lucros correspondentes.

2. **Treinamento da Rede**: Alimentamos essa rede neural com os dados coletados. Durante o treinamento, a rede ajusta os pesos de seus neurnios, tentando aproximar ao mximo a funo real que calcula o lucro a partir dos dados de entrada.

3. **Validao e Predio**: Aps o treinamento, se a rede estiver bem ajustada, ela poder prever o lucro baseado em novos dados de entrada. Este modelo pode ser extremamente til para previses financeiras e planejamento estratgico.


```python
def calcular_lucro(preco_venda, custo_total):
    lucro = preco_venda - custo_total
    return lucro
```

#### Sigmoid

Frequentemente utilizada como funo de ativao em redes neurais artificiais. Ela ajuda a introduzir no-linearidades nas sadas dos neurnios e permite que a rede neural modele relaes complexas nos dados.

A funo sigmoid  comumente representada pela funo logstica, que tem a forma matemtica:

$$ f(x) = \frac{1}{1 + e^{-x}} $$

onde $ e $  a base do logaritmo natural (aproximadamente 2.71828) e $ x $  a varivel independente. Esta funo  chamada de "sigmoid" devido  sua forma em S quando plotada em um grfico.

#### Tanh

A equao da funo de ativao **tangente hiperblica (tanh)**  dada por:

$$
\tanh(x) = \frac{e^x - e^{-x}}{e^x + e^{-x}}
$$

A funo **tanh** mapeia os valores de entrada para uma faixa entre -1 e 1, o que a torna uma funo de ativao popular em redes neurais, especialmente em redes recorrentes (RNNs). Ao contrrio da funo sigmoide, que mapeia para uma faixa entre 0 e 1, a **tanh** centra a sada em torno de zero, o que geralmente resulta em um aprendizado mais eficiente, pois a mdia dos dados pode estar mais prxima de zero.

### Propriedades principais:
- Valores de entrada grandes e positivos so mapeados para valores prximos de 1.
- Valores de entrada grandes e negativos so mapeados para valores prximos de -1.
- Valores prximos de 0 so mapeados aproximadamente linearmente entre -1 e 1.

Isso ajuda a mitigar o problema de gradientes que desaparecem em comparao com a funo sigmoide.

```python
def tanh(x):
  return (torch.exp(x) - torch.exp(-x)) / (torch.exp(x) + torch.exp(-x))
```

### ReLU, LeakyReLU, SiLU

#### ReLU

A equao da funo de ativao **ReLU (Rectified Linear Unit)**  bastante simples e pode ser expressa como:

$$
\text{ReLU}(x) = \max(0, x)
$$

**Em palavras:**
- Se o valor de entrada \(x\) for positivo, a sada  \(x\).
- Se o valor de entrada \(x\) for negativo ou zero, a sada  \(0\).

**Matematicamente:**
$$
\text{ReLU}(x) =
\begin{cases}
x, & \text{se } x > 0 \\
0, & \text{se } x \leq 0
\end{cases}
$$

A **ReLU**  amplamente usada em redes neurais profundas por ser computacionalmente eficiente e por ajudar a reduzir o problema de gradientes desaparecendo (em comparao com funes sigmoides ou tanh), uma vez que no satura para entradas positivas. No entanto, ela pode levar ao problema de "neurnios mortos" se muitos valores de entrada forem negativos, pois a derivada da ReLU  zero para entradas no positivas.

```python
def relu(x):
  return torch.where(x > 0, x, 0.0)

y_values = relu(x)
y_values.min(), y_values.max()
```

#### LeakyReLU

A equao da funo de ativao **Leaky ReLU**  uma modificao da ReLU para permitir pequenas gradientes tambm em entradas negativas. Ela  definida como:

$$
\text{Leaky ReLU}(x) =
\begin{cases}
x, & \text{se } x > 0 \\
\alpha x, & \text{se } x \leq 0
\end{cases}
$$

Onde \(\alpha\)  um pequeno valor positivo, tipicamente algo como 0.01, que define a inclinao para os valores negativos.

**Em palavras:**
- Se $ (x > 0 ) $, a sada  $ (x) $ (igual  ReLU).
- Se $ (x \leq 0) $, a sada  uma frao $ (\alpha) $ de $ (x) $, permitindo gradientes pequenos para valores negativos.

A Leaky ReLU ajuda a evitar o problema dos "neurnios mortos" que pode ocorrer com a funo ReLU pura, j que permite um pequeno gradiente mesmo quando a entrada  negativa.

```python
def leakyReLU(x):
  alpha = 0.01
  return torch.where(x > 0, x, alpha * x)
```

#### SiLU

A equao da funo de ativao **SiLU (Sigmoid Linear Unit)**, tambm conhecida como **Swish**,  dada por:

$$
\text{SiLU}(x) = x \cdot \sigma(x)
$$

Onde $ \sigma(x) $  a funo sigmoide:

$$
\sigma(x) = \frac{1}{1 + e^{-x}}
$$

### Em palavras:
A SiLU combina a entrada linear com a funo sigmoide da entrada. A multiplicao de $(x)$ pela sigmoide de $(x)$ resulta em uma ativao suave que  aproximadamente linear para valores positivos grandes e suavemente decai para valores negativos, o que pode ajudar na estabilizao e no aprendizado de redes profundas.

### Caractersticas:
- Para $(x \gg 0)$, a SiLU se aproxima de $(x)$, comportando-se como uma funo linear.
- Para $(x \ll 0)$, a SiLU se aproxima de $(0)$, suavemente.
- Ao contrrio da ReLU, ela no "mata" completamente os valores negativos, mas permite um gradiente suave.

#### Outras funes importantes

- ELU (Exponential Linear Unit)
- GELU (Gaussian Error Linear Unit)

## Otimizadores

Os otimizadores como SGD e Adam so usados em redes neurais para ajustar os parmetros (pesos) do modelo, a fim de minimizar a funo de perda durante o treinamento. O objetivo desses otimizadores  encontrar os melhores valores para os pesos que maximizem a preciso do modelo. Eles fazem isso atualizando os pesos iterativamente, baseados nas derivadas da funo de perda com relao a esses parmetros (gradientes).

#### **SGD (Stochastic Gradient Descent)**
O **SGD**  uma verso do gradiente descendente, onde em vez de usar todo o conjunto de dados para calcular os gradientes (como no gradiente descendente tradicional), ele faz atualizaes com base em pequenos lotes (mini-batches) ou at mesmo em exemplos individuais.

#### Frmula de atualizao do SGD:
$$
\theta_{t+1} = \theta_t - \eta \cdot \nabla_{\theta}J(\theta)
$$
- $ (\theta_t) $ so os parmetros do modelo no passo $ (t) $,
- $ (\eta) $  a taxa de aprendizado,
- $ \nabla_{\theta}J(\theta) $  o gradiente da funo de perda $ (J) $ com relao aos parmetros $ (\theta) $.

#### Vantagens:
- Simples de implementar.
- Rpido para grandes conjuntos de dados, j que no precisa calcular o gradiente de todo o conjunto.

#### Desvantagens:
- Pode ter oscilaes durante o treinamento, o que pode dificultar a convergncia para um timo global.
- Sensvel  escolha da taxa de aprendizado.

 w/b - lr * dw/b

```python
import torch

class SimpleSGD:
    def __init__(self, params, lr=0.01, weight_decay=0.001, momentum=0.04, dampening=0.001):
        """
        Inicializa o otimizador SGD.

        params: parmetros do modelo a serem otimizados.
        lr: taxa de aprendizado (o tamanho dos passos na direo do gradiente).
        weight_decay: valor para regularizao L2 (penalizao no tamanho dos pesos).
        momentum: frao do gradiente acumulado dos passos anteriores para acelerar a convergncia.
        dampening: amortecimento que controla a quantidade de gradiente acumulado no momentum.
        """
        self.params = list(params)  # Converte os parmetros do modelo em uma lista para fcil iterao.
        self.lr = lr  # Taxa de aprendizado.
        self.weight_decay = weight_decay  # Peso de penalizao para evitar overfitting.
        self.momentum = momentum  # Momento para suavizar as oscilaes no gradiente.
        self.dampening = dampening  # Amortecimento reduz o impacto do gradiente no momento.

        # Inicializa o buffer de momentum se o valor de momentum for maior que zero.
        if momentum != 0:
            # Para cada parmetro, cria um tensor zerado do mesmo tamanho para armazenar o momentum.
            self.state = {param: torch.zeros_like(param) for param in self.params}
        else:
            # Se o momentum for zero, o estado permanece vazio.
            self.state = {}

    def step(self):
        """
        Atualiza os parmetros do modelo em cada passo, utilizando gradientes e possveis correes
        de weight decay, momentum e dampening.
        """
        # Itera sobre todos os parmetros que foram passados ao otimizador.
        for param in self.params:
            if param.grad is None:
                # Se o parmetro no tiver gradiente (por exemplo, se no participou do clculo de perda),
                # passamos para o prximo.
                continue

            # Obtm o gradiente atual do parmetro (calculado por backpropagation).
            grad = param.grad.data

            # Aplica o weight decay (penalizao L2) se for diferente de zero.
            # Isso ajuda a evitar que os pesos cresam demais, introduzindo uma leve penalidade.
            if self.weight_decay != 0:
                grad = grad + self.weight_decay * param.data

            # Se o momentum estiver ativado, calculamos o novo gradiente com base no gradiente anterior.
            if self.momentum != 0:
                # Recupera o buffer de momentum correspondente ao parmetro atual.
                buf = self.state[param]
                # Atualiza o buffer de momentum com uma combinao do momento anterior e do gradiente atual.
                # (1 - dampening) controla o quanto do gradiente atual  adicionado.
                buf.mul_(self.momentum).add_(grad, alpha=(1 - self.dampening))

                # O gradiente final a ser aplicado  o valor ajustado pelo momentum.
                grad = buf

            # Atualiza os parmetros aplicando o gradiente (ajustado pelo momentum e pelo weight decay).
            # A taxa de aprendizado (lr) multiplica o gradiente, determinando o tamanho do ajuste.
            param.data = param.data - self.lr * grad

    def zero_grad(self):
        """
        Zera os gradientes de todos os parmetros.
        Isso  necessrio antes de recalcular os gradientes no prximo passo de treinamento.
        """
        for param in self.params:
            if param.grad is not None:
                # Detach o gradiente para evitar problemas de computao com autograd
                param.grad.detach_()
                # Zera os gradientes (necessrio para no acumular gradientes de diferentes passos).
                param.grad.zero_()
```

```python
model = Model()
train = Train(model, nn.MSELoss(), SimpleSGD(model.parameters(), lr=0.0001))
train.train_step(300, train_dataloader, test_dataloader)
```
