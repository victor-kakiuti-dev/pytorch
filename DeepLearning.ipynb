###📏 Função Linear

Uma função linear é uma função matemática que pode ser expressa na forma $f(x) = wx + b$, onde $x$ representa a variável independente, $w$ é a inclinação da reta (também conhecida como coeficiente angular), e $b$ é o intercepto $y$, ou seja, o ponto em que a reta cruza o eixo $y$ no plano cartesiano. A principal característica de uma função linear é que ela cria uma linha reta quando gerado um gráfico.

Aqui estão algumas características importantes das funções lineares:

- **Inclinação ou coeficiente angular ($w$):** Determina a inclinação da linha. Se $w$ for positivo, a linha inclina-se para cima à medida que se move da esquerda para a direita. Se $w$ for negativo, a linha inclina-se para baixo.
- **Intercepto y ($b$):** É o ponto onde a linha cruza o eixo y.

###🧠 Aprendizado

image.png

**Processo de aprendizado:**

1. **Propagação para frente (Forward Propagation) 🚀**: Onde as entradas são passadas através da rede, através de suas camadas ocultas até a camada de saída para obter a saída prevista.

2. **Cálculo do Erro 🎯**: A saída prevista é comparada com a saída real e calcula-se a diferença ou erro, geralmente utilizando uma função de perda como a entropia cruzada ou o erro quadrático médio.

3. **Backpropagation 🔙**: Neste passo, o erro é propagado de volta pela rede, começando da saída para as camadas anteriores. O gradiente da função de perda é calculado em relação a cada peso, o que indica como os pesos devem ser ajustados para reduzir o erro.

4. **Atualização dos Pesos ⚖️**: Com o gradiente calculado, os pesos são ajustados.

```python
import torch

w = torch.tensor(0.3, requires_grad=True)
b = torch.tensor(-0.01, requires_grad=True)
x = torch.arange(-10,10)
y = 3.2 *  torch.arange(-10,10) + 3
```

```python
def perceptron(w,x,b):
  return w * x + b
```

```python
w_values, b_values = [], []
for epoch in range(200):

  #Etapa do feed Forward
  w_values.append(w.item())
  b_values.append(b.item())
  output = perceptron(w,x,b)

  #Calcular o erro
  squared_error = (output - y)**2
  error = torch.sum(squared_error)

  print(f"Epoch: {epoch}, Error: {error.item()}")

  #Calcular a derivada parcial
  error.backward()

  # Ajusta os valores de w e b
  with torch.no_grad():
    w -= 0.001 * w.grad
    b -= 0.001 * b.grad

  w.grad.zero_()
  b.grad.zero_()
```

```python
output = perceptron(w,x,b)
output, y
```

###📉 funções

Na matemática, uma função é um conceito fundamental que descreve uma relação especial entre conjuntos de números ou mais geralmente, entre elementos de conjuntos quaisquer. Uma função associa dois conjuntos de elementos de tal forma que a cada elemento do primeiro conjunto, chamado de domínio, corresponde exatamente um elemento no segundo conjunto, chamado de contradomínio ou imagem.

### Características Principais de uma Função:

1. **Domínio**: O conjunto de todas as entradas possíveis para a função. Por exemplo, o conjunto de todos os números reais, todos os números inteiros, ou qualquer outro conjunto definido.

2. **Contradomínio**: O conjunto de todas as saídas possíveis que a função pode produzir. Nem todas as saídas possíveis precisam ser efetivamente atingidas pela função; as que são atingidas são chamadas de imagem.

3. **Regra de Correspondência**: Uma "regra" que define como cada entrada é associada a uma saída. Essa regra pode ser uma expressão matemática, uma equação, ou uma descrição verbal que pode ser sistematicamente aplicada para determinar a saída a partir de uma entrada.

#### Tipos de Funções:

- **Funções Lineares**: Funções que formam uma linha reta quando graficadas, como $ f(x) = 2x + 3 $.
- **Funções Polinomiais**: Incluem termos de potências de $ x $, como $ f(x) = x^2 - 4x + 4 $.
- **Funções Trigonométricas**: Envolve funções seno, cosseno, tangente, etc., como $ f(x) = sin(x)$.
- **Funções Exponenciais e Logarítmicas**: Involvem exponenciação e logaritmos, como $ f(x) = e^x $ ou $ f(x) = \log(x) $.

#### Importância das Funções:

Funções são fundamentais em todas as áreas da matemática e suas aplicações, incluindo física, engenharia, economia, biologia, e muitas outras ciências. Elas ajudam a modelar relações e processos do mundo real, permitindo previsões e compreensões profundas de fenômenos naturais e humanos.

Em resumo, uma função é uma maneira de expressar uma dependência entre quantidades, onde a variação de uma quantidade determina a variação de outra.

###📈 Derivada

A derivada de uma função em um determinado ponto é a inclinação da reta tangente à função nesse ponto. Em termos mais formais, a derivada de uma função $ f(x) $ em relação a $ x $ é definida como o limite da razão entre a mudança de $ f(x) $ e a mudança de $ x $ quando $ x $ tende a zero. A expressão matemática para a derivada é:

$ f'(x) = \lim_{\Delta x \to 0} \frac{f(x + \Delta x) - f(x)}{\Delta x} $

Esta expressão calcula a taxa instantânea de mudança da função $ f(x) $ em relação a $ x $. Em outras palavras, nos dá a inclinação da reta tangente ao gráfico da função no ponto $ x $. A reta tangente é a reta que "apenas toca" o gráfico da função em um único ponto, e a inclinação dessa reta tangente é um indicador da rapidez com que a função está aumentando ou diminuindo nesse ponto.

###🔄 **Regra da cadeia**

A regra da cadeia é uma técnica fundamental no cálculo diferencial que é utilizada para derivar funções compostas. Essencialmente, ela permite calcular a derivada de uma função que é o resultado da composição de duas ou mais funções.

#### Forma Normal (Propagação Direta da Derivada) 1 variavel

Neste método, as derivadas são calculadas diretamente de acordo com a composição das funções.

Vamos usar as funções $ b(x) $ e $ n(x) $ e aplicar a regra da cadeia para calcular a derivada de uma função composta utilizando esses nomes de função.

**Definição das Funções**

Suponhamos que:

- $ b(x) $ seja uma função de $ x $.
- $ n(x) $ seja outra função de $ x $.
- Queremos derivar uma função composta $ f(x) = b(n(x)) $.

**Aplicação da Regra da Cadeia**

1. **Identificar as funções:**
   - $ n(x) $ é a função interna.
   - $ b(u) $ é a função externa, onde $ u = n(x) $.

2. **Derivar cada função:**
   - Derive $ n(x) $ em relação a $ x $: $ n'(x) $.
   - Derive $ b(u) $ em relação a $ u $: $ b'(u) $.

3. **Aplicar a regra da cadeia:**
   - A derivada da função composta $ f(x) $ é dada por:
$
     f'(x) = b'(n(x)) \cdot n'(x)
$
     

**Exemplo Prático**

Suponha que:

- $ n(x) = x^2 $ (a função interna).
- $ b(u) = \sin(u) $ (a função externa).

Então:

1. **Função Composta:** $ f(x) = \sin(x^2) $.
2. **Derivada da função interna $ n(x) $:**
   - $ n'(x) = 2x $.
3. **Derivada da função externa $ b(u) $ quando $ u = x^2 $:**
   - $ b'(u) = \cos(u) $.
4. **Aplicando a regra da cadeia:**
   - $ f'(x) = \cos(x^2) \cdot 2x $.

###🏞️ Descida do Gradiente

A derivada parcial de uma função em relação a uma variável indica a taxa de variação de $ f $ em relação àquela variável, mantendo as outras constantes. O sinal da derivada parcial pode te dar informações importantes:

- Se a derivada parcial é positiva ($ + $), isso indica que um pequeno aumento na variável leva a um aumento no valor de $ f $. Ou seja, $ f $ cresce à medida que a variável aumenta.
- Se a derivada parcial é negativa ($ - $), isso significa que um pequeno aumento na variável leva a uma diminuição no valor de $ f $. Portanto, $ f $ diminui à medida que a variável aumenta.

#### O que é o Gradiente?
O gradiente de uma função $ f(x_1, x_2, \ldots, x_n) $ em um ponto específico é um vetor que contém todas as derivadas parciais da função naquele ponto. Formalmente, é definido como:
$$
\nabla f = \left( \frac{\partial f}{\partial x_1}, \frac{\partial f}{\partial x_2}, \ldots, \frac{\partial f}{\partial x_n} \right)
$$
Este vetor não apenas indica a direção de maior aumento, mas sua magnitude também fornece uma noção da taxa de aumento da função nessa direção.

Geometricamente, você pode imaginar o gradiente como uma seta que aponta para cima a partir de um ponto em uma superfície montanhosa. Se você seguir na direção dessa seta, estará subindo a montanha pelo caminho mais íngreme.

Quando todas essas mudanças são combinadas (como no vetor gradiente), elas apontam para a direção onde a função cresce mais rapidamente se todas as variáveis forem alteradas proporcionalmente ao valor de suas respectivas derivadas parciais.

### Descida do Gradiente

Na técnica de otimização conhecida como "descida do gradiente", o objetivo é encontrar o minímo de uma função. Para fazer isso, utilizamos o fato de que o gradiente aponta para o maior aumento e, portanto, o oposto do gradiente (ou gradiente negativo) apontará para a maior diminuição. Isso significa que, se movermos na direção oposta ao gradiente, estamos movendo na direção que reduz o valor da função mais rapidamente.


```python
import numpy as np
import matplotlib.pyplot as plt

# Função e gradiente
def f(x):
    return x**2 - 4*x + 4

def grad_f(x):
    return 2*x - 4

# Descida do gradiente
def gradient_descent(initial_x, learning_rate, n_steps):
    x = initial_x
    trajectory = [x]
    for _ in range(n_steps):
        gradient = grad_f(x)
        x = x - learning_rate * gradient
        trajectory.append(x)
    return trajectory

# Gerar dados
x_values = np.linspace(-1, 5, 400)
y_values = f(x_values)
trajectory = gradient_descent(0.5, 0.1, 15)

# Gerar o gráfico
plt.figure(figsize=(8, 5))
plt.plot(x_values, y_values, label='f(x) = x² - 4x + 4')
plt.scatter(trajectory, f(np.array(trajectory)), color='red', s=50, label='Descida do Gradiente')
plt.plot(trajectory, f(np.array(trajectory)), color='red', linestyle='--', alpha=0.5)
plt.title('Exemplo de Descida do Gradiente')
plt.xlabel('x')
plt.ylabel('f(x)')
plt.legend()
plt.grid(True)
plt.show()
```

##🌐 Teorema da Aproximação Universal

**O que é o Teorema da Aproximação Universal?**

O Teorema da Aproximação Universal afirma que uma rede neural com uma única camada oculta pode aproximar qualquer função contínua em espaços compactos. Isso significa que, com a arquitetura adequada, uma rede neural pode teoricamente modelar qualquer relação matemática, não importando sua complexidade.

**Aplicação Prática: Modelando o Lucro de uma Empresa**

Suponha que desejamos entender ou prever o lucro de uma empresa com base em fatores como vendas, custos e impostos. Utilizando o teorema da aproximação universal, podemos configurar uma rede neural para aprender a função que liga esses inputs ao lucro.

1. **Coleta de Dados**: Primeiramente, coletamos dados históricos que incluem vendas, custos, e outros fatores relevantes junto com os lucros correspondentes.

2. **Treinamento da Rede**: Alimentamos essa rede neural com os dados coletados. Durante o treinamento, a rede ajusta os pesos de seus neurônios, tentando aproximar ao máximo a função real que calcula o lucro a partir dos dados de entrada.

3. **Validação e Predição**: Após o treinamento, se a rede estiver bem ajustada, ela poderá prever o lucro baseado em novos dados de entrada. Este modelo pode ser extremamente útil para previsões financeiras e planejamento estratégico.


```python
def calcular_lucro(preco_venda, custo_total):
    lucro = preco_venda - custo_total
    return lucro
```

### Dataset (California Housing Prices)

Preços médios de casas nos distritos da Califórnia derivados do censo de 1990.

https://www.kaggle.com/datasets/camnugent/california-housing-prices

```python
from sklearn.preprocessing import RobustScaler
from sklearn.model_selection import train_test_split
import torch
import torch.nn as nn
from torch.utils.data import Dataset, DataLoader
from sklearn.datasets import fetch_california_housing

# Carregar dataset
data = fetch_california_housing()

# Aplicar escalonamento robusto
scaler = RobustScaler()
features = scaler.fit_transform(data['data'])

# Separar dados em treinamento e teste (80% treino, 20% teste)
features_train, features_test, targets_train, targets_test = train_test_split(
    features, data['target'], test_size=0.2, random_state=42
)

# Definir Dataset customizado no PyTorch
class CaliforniaHousingDataset(Dataset):
    def __init__(self, features, targets):
        self.features = torch.tensor(features, dtype=torch.float32)
        self.targets = torch.tensor(targets, dtype=torch.float32).unsqueeze(1)

    def __len__(self):
        return len(self.features)

    def __getitem__(self, idx):
        return self.features[idx], self.targets[idx]

# Criar datasets de treinamento e teste
train_dataset = CaliforniaHousingDataset(features_train, targets_train)
test_dataset = CaliforniaHousingDataset(features_test, targets_test)

# Criar os dataloaders
train_dataloader = DataLoader(train_dataset, batch_size=32, shuffle=True)
test_dataloader = DataLoader(test_dataset, batch_size=32, shuffle=False)

# Verificar o primeiro batch de treinamento
first_batch = next(iter(train_dataloader))
first_batch[0].shape, first_batch[1].shape
```

### Definição do Modelo

```python
class Linear(nn.Module):
  def __init__(self, in_features, out_features):
    super(Linear, self).__init__()
    self.w = nn.Parameter(torch.randn(in_features, out_features), requires_grad=True)
    self.b = nn.Parameter(torch.zeros(out_features), requires_grad=True)

  def forward(self, x):
    return torch.matmul(x, self.w) + self.b
```

```python
x = torch.rand(10, 8)
x.shape
```

```python
linear = Linear(8, 20)
linear(x).shape
```

```python
class Model(nn.Module):
  def __init__(self):
    super(Model, self).__init__()
    self.linear1 = Linear(8, 6)
    self.linear2 = Linear(6, 3)
    self.linear3 = Linear(3, 1)

  def forward(self, x):
    x = self.linear1(x)
    x = torch.relu(x)
    x = self.linear2(x)
    x = torch.relu(x)
    x = self.linear3(x)
    return x
```

```python
model = Model()

x = torch.rand(32, 8)
model(x).shape
```

### Treinamento do Modelo

```python
class Train:
  def __init__(self, model, loss_fn, optimizer):
    self.model = model
    self.loss_fn = loss_fn
    self.optimizer = optimizer

  def train_step(self, epochs, train_dataloader, test_dataloader):

    for epoch in range(epochs):

      train_loss = []
      test_loss = []
      for batch in train_dataloader:
        x, y = batch

        self.optimizer.zero_grad()
        y_pred = self.model(x)
        loss = self.loss_fn(y_pred, y)
        loss.backward()
        self.optimizer.step()
        train_loss.append(loss.item())

      for batch in test_dataloader:
        with torch.no_grad():
          x, y = batch
          y_pred = self.model(x)
          loss = self.loss_fn(y_pred, y)
          test_loss.append(loss.item())

      train_loss = torch.mean(torch.tensor(train_loss))
      test_loss = torch.mean(torch.tensor(test_loss))
      print(f"Epoch: {epoch + 1}/{epochs}, Train Loss: {train_loss}, Test Loss: {test_loss}")
```

```python
model = Model()
train = Train(model, nn.MSELoss(), torch.optim.Adam(model.parameters(), lr=0.0001))
train.train_step(300, train_dataloader, test_dataloader)
```

## Funções de ativação

### Sigmoid e Tanh

#### Sigmoid

Frequentemente utilizada como função de ativação em redes neurais artificiais. Ela ajuda a introduzir não-linearidades nas saídas dos neurônios e permite que a rede neural modele relações complexas nos dados.

A função sigmoid é comumente representada pela função logística, que tem a forma matemática:

$$ f(x) = \frac{1}{1 + e^{-x}} $$

onde $ e $ é a base do logaritmo natural (aproximadamente 2.71828) e $ x $ é a variável independente. Esta função é chamada de "sigmoid" devido à sua forma em S quando plotada em um gráfico.

https://www.sciencedirect.com/topics/computer-science/sigmoid-function

```python
import torch
import numpy as np
import matplotlib.pyplot as plt

def show_chart(x_values, y_values, label):
  plt.figure(figsize=(8, 6))
  plt.plot(x_values, y_values, label=label, color='b')
  plt.title(f'Gráfico da Função {label}')
  plt.xlabel('x')
  plt.ylabel(f'{label}(x)')
  plt.grid(True)
  plt.legend()
  plt.show()
```

#### Tanh

A equação da função de ativação **tangente hiperbólica (tanh)** é dada por:

$$
\tanh(x) = \frac{e^x - e^{-x}}{e^x + e^{-x}}
$$

A função **tanh** mapeia os valores de entrada para uma faixa entre -1 e 1, o que a torna uma função de ativação popular em redes neurais, especialmente em redes recorrentes (RNNs). Ao contrário da função sigmoide, que mapeia para uma faixa entre 0 e 1, a **tanh** centra a saída em torno de zero, o que geralmente resulta em um aprendizado mais eficiente, pois a média dos dados pode estar mais próxima de zero.

### Propriedades principais:
- Valores de entrada grandes e positivos são mapeados para valores próximos de 1.
- Valores de entrada grandes e negativos são mapeados para valores próximos de -1.
- Valores próximos de 0 são mapeados aproximadamente linearmente entre -1 e 1.

Isso ajuda a mitigar o problema de gradientes que desaparecem em comparação com a função sigmoide.

```python
def tanh(x):
  return (torch.exp(x) - torch.exp(-x)) / (torch.exp(x) + torch.exp(-x))
```

```python
y_values = tanh(x)
y_values.min(), y_values.max()
```

```python
show_chart(x, y_values, "tanh")
```

### ReLU, LeakyReLU, SiLU

#### ReLU

A equação da função de ativação **ReLU (Rectified Linear Unit)** é bastante simples e pode ser expressa como:

$$
\text{ReLU}(x) = \max(0, x)
$$

**Em palavras:**
- Se o valor de entrada \(x\) for positivo, a saída é \(x\).
- Se o valor de entrada \(x\) for negativo ou zero, a saída é \(0\).

**Matematicamente:**
$$
\text{ReLU}(x) =
\begin{cases}
x, & \text{se } x > 0 \\
0, & \text{se } x \leq 0
\end{cases}
$$

A **ReLU** é amplamente usada em redes neurais profundas por ser computacionalmente eficiente e por ajudar a reduzir o problema de gradientes desaparecendo (em comparação com funções sigmoides ou tanh), uma vez que não satura para entradas positivas. No entanto, ela pode levar ao problema de "neurônios mortos" se muitos valores de entrada forem negativos, pois a derivada da ReLU é zero para entradas não positivas.

```python
def relu(x):
  return torch.where(x > 0, x, 0.0)

y_values = relu(x)
y_values.min(), y_values.max()
```

```python
relu = nn.ReLU()
relu(x)
```

#### LeakyReLU

A equação da função de ativação **Leaky ReLU** é uma modificação da ReLU para permitir pequenas gradientes também em entradas negativas. Ela é definida como:

$$
\text{Leaky ReLU}(x) =
\begin{cases}
x, & \text{se } x > 0 \\
\alpha x, & \text{se } x \leq 0
\end{cases}
$$

Onde \(\alpha\) é um pequeno valor positivo, tipicamente algo como 0.01, que define a inclinação para os valores negativos.

**Em palavras:**
- Se $ (x > 0 ) $, a saída é $ (x) $ (igual à ReLU).
- Se $ (x \leq 0) $, a saída é uma fração $ (\alpha) $ de $ (x) $, permitindo gradientes pequenos para valores negativos.

A Leaky ReLU ajuda a evitar o problema dos "neurônios mortos" que pode ocorrer com a função ReLU pura, já que permite um pequeno gradiente mesmo quando a entrada é negativa.

```python
def leakyReLU(x):
  alpha = 0.01
  return torch.where(x > 0, x, alpha * x)
```

```python
y_values = leakyReLU(x)
y_values.min(), y_values.max()
```

```python
leaky_relu = nn.LeakyReLU()
leaky_relu(x)
```

```python
show_chart(x, y_values, "LeakyReLU")
```

#### SiLU

A equação da função de ativação **SiLU (Sigmoid Linear Unit)**, também conhecida como **Swish**, é dada por:

$$
\text{SiLU}(x) = x \cdot \sigma(x)
$$

Onde $ \sigma(x) $ é a função sigmoide:

$$
\sigma(x) = \frac{1}{1 + e^{-x}}
$$

### Em palavras:
A SiLU combina a entrada linear com a função sigmoide da entrada. A multiplicação de $(x)$ pela sigmoide de $(x)$ resulta em uma ativação suave que é aproximadamente linear para valores positivos grandes e suavemente decai para valores negativos, o que pode ajudar na estabilização e no aprendizado de redes profundas.

### Características:
- Para $(x \gg 0)$, a SiLU se aproxima de $(x)$, comportando-se como uma função linear.
- Para $(x \ll 0)$, a SiLU se aproxima de $(0)$, suavemente.
- Ao contrário da ReLU, ela não "mata" completamente os valores negativos, mas permite um gradiente suave.

#### Outras funções importantes
- Softmax
- ELU (Exponential Linear Unit)
- GELU (Gaussian Error Linear Unit)

https://pytorch.org/docs/stable/nn.html#non-linear-activations-weighted-sum-nonlinearity

## Otimizadores

Os otimizadores como SGD e Adam são usados em redes neurais para ajustar os parâmetros (pesos) do modelo, a fim de minimizar a função de perda durante o treinamento. O objetivo desses otimizadores é encontrar os melhores valores para os pesos que maximizem a precisão do modelo. Eles fazem isso atualizando os pesos iterativamente, baseados nas derivadas da função de perda com relação a esses parâmetros (gradientes).

#### **SGD (Stochastic Gradient Descent)**
O **SGD** é uma versão do gradiente descendente, onde em vez de usar todo o conjunto de dados para calcular os gradientes (como no gradiente descendente tradicional), ele faz atualizações com base em pequenos lotes (mini-batches) ou até mesmo em exemplos individuais.

#### Fórmula de atualização do SGD:
$$
\theta_{t+1} = \theta_t - \eta \cdot \nabla_{\theta}J(\theta)
$$
- $ (\theta_t) $ são os parâmetros do modelo no passo $ (t) $,
- $ (\eta) $ é a taxa de aprendizado,
- $ \nabla_{\theta}J(\theta) $ é o gradiente da função de perda $ (J) $ com relação aos parâmetros $ (\theta) $.

#### Vantagens:
- Simples de implementar.
- Rápido para grandes conjuntos de dados, já que não precisa calcular o gradiente de todo o conjunto.

#### Desvantagens:
- Pode ter oscilações durante o treinamento, o que pode dificultar a convergência para um ótimo global.
- Sensível à escolha da taxa de aprendizado.

 w/b - lr * dw/b

```python
import torch

class SimpleSGD:
    def __init__(self, params, lr=0.01, weight_decay=0.001, momentum=0.04, dampening=0.001):
        """
        Inicializa o otimizador SGD.

        params: parâmetros do modelo a serem otimizados.
        lr: taxa de aprendizado (o tamanho dos passos na direção do gradiente).
        weight_decay: valor para regularização L2 (penalização no tamanho dos pesos).
        momentum: fração do gradiente acumulado dos passos anteriores para acelerar a convergência.
        dampening: amortecimento que controla a quantidade de gradiente acumulado no momentum.
        """
        self.params = list(params)  # Converte os parâmetros do modelo em uma lista para fácil iteração.
        self.lr = lr  # Taxa de aprendizado.
        self.weight_decay = weight_decay  # Peso de penalização para evitar overfitting.
        self.momentum = momentum  # Momento para suavizar as oscilações no gradiente.
        self.dampening = dampening  # Amortecimento reduz o impacto do gradiente no momento.

        # Inicializa o buffer de momentum se o valor de momentum for maior que zero.
        if momentum != 0:
            # Para cada parâmetro, cria um tensor zerado do mesmo tamanho para armazenar o momentum.
            self.state = {param: torch.zeros_like(param) for param in self.params}
        else:
            # Se o momentum for zero, o estado permanece vazio.
            self.state = {}

    def step(self):
        """
        Atualiza os parâmetros do modelo em cada passo, utilizando gradientes e possíveis correções
        de weight decay, momentum e dampening.
        """
        # Itera sobre todos os parâmetros que foram passados ao otimizador.
        for param in self.params:
            if param.grad is None:
                # Se o parâmetro não tiver gradiente (por exemplo, se não participou do cálculo de perda),
                # passamos para o próximo.
                continue

            # Obtém o gradiente atual do parâmetro (calculado por backpropagation).
            grad = param.grad.data

            # Aplica o weight decay (penalização L2) se for diferente de zero.
            # Isso ajuda a evitar que os pesos cresçam demais, introduzindo uma leve penalidade.
            if self.weight_decay != 0:
                grad = grad + self.weight_decay * param.data

            # Se o momentum estiver ativado, calculamos o novo gradiente com base no gradiente anterior.
            if self.momentum != 0:
                # Recupera o buffer de momentum correspondente ao parâmetro atual.
                buf = self.state[param]
                # Atualiza o buffer de momentum com uma combinação do momento anterior e do gradiente atual.
                # (1 - dampening) controla o quanto do gradiente atual é adicionado.
                buf.mul_(self.momentum).add_(grad, alpha=(1 - self.dampening))

                # O gradiente final a ser aplicado é o valor ajustado pelo momentum.
                grad = buf

            # Atualiza os parâmetros aplicando o gradiente (ajustado pelo momentum e pelo weight decay).
            # A taxa de aprendizado (lr) multiplica o gradiente, determinando o tamanho do ajuste.
            param.data = param.data - self.lr * grad

    def zero_grad(self):
        """
        Zera os gradientes de todos os parâmetros.
        Isso é necessário antes de recalcular os gradientes no próximo passo de treinamento.
        """
        for param in self.params:
            if param.grad is not None:
                # Detach o gradiente para evitar problemas de computação com autograd
                param.grad.detach_()
                # Zera os gradientes (necessário para não acumular gradientes de diferentes passos).
                param.grad.zero_()
```

```python
model = Model()
train = Train(model, nn.MSELoss(), SimpleSGD(model.parameters(), lr=0.0001))
train.train_step(300, train_dataloader, test_dataloader)
```

Para normalizar valores entre 0 e 1, você pode usar a fórmula de **min-max normalization**:

$$
X_{\text{normalizado}} = \frac{X - X_{\text{min}}}{X_{\text{max}} - X_{\text{min}}}
$$

```python
from sklearn.model_selection import train_test_split
import torch
import torch.nn as nn
from torch.utils.data import Dataset, DataLoader
from sklearn.datasets import load_breast_cancer

def normalize(x):
  return (x - x.min()) / (x.max() - x.min())

# Carregar dataset
data = load_breast_cancer()

# Aplicar escalonamento robusto
features = normalize(data['data'])

# Separar dados em treinamento e teste (80% treino, 20% teste)
features_train, features_test, targets_train, targets_test = train_test_split(
    features, data['target'], test_size=0.2, random_state=42
)

# Definir Dataset customizado no PyTorch
class CaliforniaHousingDataset(Dataset):
    def __init__(self, features, targets):
        self.features = torch.tensor(features, dtype=torch.float32)
        self.targets = torch.tensor(targets, dtype=torch.float32).unsqueeze(1)

    def __len__(self):
        return len(self.features)

    def __getitem__(self, idx):
        return self.features[idx], self.targets[idx]

# Criar datasets de treinamento e teste
train_dataset = CaliforniaHousingDataset(features_train, targets_train)
test_dataset = CaliforniaHousingDataset(features_test, targets_test)

# Criar os dataloaders
train_dataloader = DataLoader(train_dataset, batch_size=32, shuffle=True)
test_dataloader = DataLoader(test_dataset, batch_size=32, shuffle=False)

# Verificar o primeiro batch de treinamento
first_batch = next(iter(train_dataloader))
first_batch[0].shape, first_batch[1].shape
```

```python
class Linear(nn.Module):
  def __init__(self, in_features, out_features):
    super(Linear, self).__init__()
    self.w = nn.Parameter(torch.randn(in_features, out_features), requires_grad=True)
    self.b = nn.Parameter(torch.zeros(out_features), requires_grad=True)

  def forward(self, x):
    return torch.matmul(x, self.w) + self.b

class Model(nn.Module):
  def __init__(self):
    super(Model, self).__init__()
    self.linear1 = Linear(30, 26)
    self.linear2 = Linear(26, 16)
    self.linear3 = Linear(16, 6)
    self.linear4 = Linear(6, 1)

  def forward(self, x):
    x = self.linear1(x)
    x = silu(x)
    x = self.linear2(x)
    x = silu(x)
    x = self.linear3(x)
    x = silu(x)
    x = self.linear4(x)
    x = torch.sigmoid(x)
    return x

class Train:
  def __init__(self, model, loss_fn, optimizer):
    self.model = model
    self.loss_fn = loss_fn
    self.optimizer = optimizer

  def train_step(self, epochs, train_dataloader, test_dataloader):

    for epoch in range(epochs):

      train_loss = []
      test_loss = []
      for batch in train_dataloader:
        x, y = batch

        self.optimizer.zero_grad()
        y_pred = self.model(x)
        loss = self.loss_fn(y_pred, y)
        loss.backward()
        self.optimizer.step()
        train_loss.append(loss.item())

      for batch in test_dataloader:
        with torch.no_grad():
          x, y = batch
          y_pred = self.model(x)
          loss = self.loss_fn(y_pred, y)
          test_loss.append(loss.item())

      train_loss = torch.mean(torch.tensor(train_loss))
      test_loss = torch.mean(torch.tensor(test_loss))

      if epoch % 10 == 0:
          print(f"Epoch: {epoch + 1}/{epochs}, Train Loss: {train_loss}, Test Loss: {test_loss}")


model = Model()
train = Train(model, nn.BCELoss(), torch.optim.Adam(model.parameters(), lr=0.0001))
train.train_step(1000, train_dataloader, test_dataloader)
```
