###üì° Broadcasting

Broadcasting √© um mecanismo que permite realizar opera√ß√µes aritm√©ticas entre tensores de diferentes formas (ou tamanhos). Vamos quebrar o texto em partes para entender melhor:

1. **"Cada tensor possui pelo menos uma dimens√£o."**
   - Isso significa que os tensores envolvidos na opera√ß√£o devem ter pelo menos uma dimens√£o. Um tensor n√£o pode ser um escalar (um √∫nico n√∫mero sem dimens√µes) se quisermos aplicar as regras de broadcasting.

2. **"Ao iterar sobre os tamanhos das dimens√µes, come√ßando pela dimens√£o final, os tamanhos das dimens√µes devem ser iguais, um deles √© 1, ou um deles n√£o existe."**
   - Esta regra √© aplicada comparando as dimens√µes dos dois tensores, come√ßando pela √∫ltima dimens√£o (a mais √† direita) e movendo-se para a primeira dimens√£o (a mais √† esquerda).
   - Para cada par de dimens√µes comparadas, tr√™s cen√°rios permitem o broadcasting:
     1. **Os tamanhos das dimens√µes s√£o iguais:** Se as dimens√µes correspondentes nos dois tensores t√™m o mesmo tamanho, ent√£o elas s√£o compat√≠veis.
     2. **Uma das dimens√µes √© 1:** Se um tensor tem uma dimens√£o de tamanho 1 na posi√ß√£o que est√° sendo comparada, ele pode ser "esticado" ou "expandido" para corresponder ao tamanho da outra dimens√£o. Isso √© feito sem realmente copiar dados, mas sim adaptando a opera√ß√£o aritm√©tica para se comportar como se o tensor menor tivesse sido expandido.
     3. **Uma das dimens√µes n√£o existe:** Se um dos tensores tem menos dimens√µes que o outro, podemos considerar que ele tem dimens√µes extras de tamanho 1 na frente (mais √† esquerda). Essas dimens√µes "faltantes" s√£o implicitamente consideradas como 1, permitindo o broadcasting.

#### **torch.reshape() e torch.view()**

No PyTorch, `torch.reshape()` e `torch.view()` s√£o duas fun√ß√µes usadas para modificar a forma de tensores, mas elas t√™m diferen√ßas sutis no seu comportamento e uso:

1. **`torch.reshape(tensor, shape)`**

- **Flexibilidade**: `torch.reshape()` pode ser usado para alterar a forma de um tensor para qualquer outra forma, desde que o n√∫mero total de elementos seja o mesmo. √â mais flex√≠vel porque, internamente, pode copiar dados para um novo tensor se necess√°rio.
- **C√≥pia vs. Vista**: Dependendo da disposi√ß√£o original dos dados na mem√≥ria, `torch.reshape()` pode retornar uma vista (um novo tensor que compartilha dados com o tensor original) ou uma c√≥pia do tensor original. Isso significa que, se voc√™ modificar os dados em um tensor retornado por `reshape()`, essas modifica√ß√µes podem ou n√£o ser refletidas no tensor original, dependendo de se uma vista foi retornada ou n√£o.

2. **`torch.view(tensor, shape)`**

- **Exig√™ncia de Contiguidade**: `torch.view()` exige que o tensor original seja cont√≠guo na mem√≥ria (ou seja, os elementos s√£o armazenados em sequ√™ncia sem lacunas) para poder retornar uma vista com a nova forma desejada. Se o tensor original n√£o for cont√≠guo, voc√™ receber√° um erro ao tentar usar `view()`. Nesse caso, voc√™ pode usar `tensor.contiguous()` antes de chamar `view()`.
- **Sempre Retorna uma Vista**: Diferente de `reshape()`, `view()` sempre retorna uma view do tensor original, n√£o uma c√≥pia. Isso significa que qualquer modifica√ß√£o nos dados do tensor retornado por `view()` ser√° refletida no tensor original, e vice-versa.

3. Escolha entre `torch.reshape()` e `torch.view()`

A escolha entre `torch.reshape()` e `torch.view()` geralmente depende de dois fatores: se voc√™ precisa garantir que o tensor retornado seja uma vista e n√£o uma c√≥pia (caso em que voc√™ usaria `view()`), e se o tensor original √© cont√≠guo na mem√≥ria (caso em que voc√™ pode usar `view()` sem problemas). Se n√£o tiver certeza sobre a contiguidade do tensor ou se precisar de flexibilidade para lidar com tensores n√£o cont√≠guos, `torch.reshape()` √© a escolha mais segura.

#### **torch.squeeze() e torch.unsqueeze()**

`torch.squeeze()` e `torch.unsqueeze()` s√£o duas fun√ß√µes complementares no PyTorch que s√£o usadas para modificar as dimens√µes de um tensor, mas elas servem para prop√≥sitos opostos:

1. **`torch.squeeze(input, dim=None)`**

- **Prop√≥sito**: A fun√ß√£o `torch.squeeze()` √© usada para remover as dimens√µes de tamanho 1 do tensor. Por exemplo, se um tensor tem a forma (2, 1, 2, 1, 2), o uso de `squeeze()` resultar√° em uma forma de (2, 2, 2), removendo as dimens√µes intermedi√°rias que tinham tamanho 1.
- **Par√¢metro `dim`**: Quando o par√¢metro `dim` n√£o √© especificado, `torch.squeeze()` remove todas as dimens√µes de tamanho 1 do tensor. Se `dim` for especificado, a fun√ß√£o tentar√° remover a dimens√£o apenas na posi√ß√£o especificada, se aquela dimens√£o tiver tamanho 1. Se a dimens√£o especificada n√£o for de tamanho 1, o tensor ser√° retornado sem altera√ß√µes.
- **Uso**: √â √∫til para remover dimens√µes redundantes ou desnecess√°rias de um tensor, por exemplo, ao receber a sa√≠da de uma camada que introduz uma dimens√£o extra de tamanho 1 que n√£o √© necess√°ria para as opera√ß√µes subsequentes.

2. **`torch.unsqueeze(input, dim)`**

- **Prop√≥sito**: A fun√ß√£o `torch.unsqueeze()` √© usada para adicionar uma dimens√£o de tamanho 1 ao tensor em uma posi√ß√£o especificada. Por exemplo, se um tensor tem a forma (3, 4), ao usar `unsqueeze()` com `dim=0`, a nova forma ser√° (1, 3, 4), adicionando uma dimens√£o extra no in√≠cio.
- **Par√¢metro `dim`**: O par√¢metro `dim` especifica a posi√ß√£o na qual a nova dimens√£o ser√° inserida. Ele √© obrigat√≥rio para `torch.unsqueeze()`, pois voc√™ precisa definir onde adicionar a nova dimens√£o.
- **Uso**: √â √∫til quando voc√™ precisa ajustar a forma de um tensor para compatibilidade de opera√ß√µes, como preparar um tensor para uma opera√ß√£o de broadcasting ou atender √† expectativa de entrada de uma camada espec√≠fica em uma rede neural.

Em resumo, `torch.squeeze()` √© usado para remover dimens√µes de tamanho 1, simplificando a estrutura do tensor, enquanto `torch.unsqueeze()` √© usado para adicionar uma nova dimens√£o de tamanho 1, permitindo manipular a forma do tensor para adequ√°-lo √†s necessidades de opera√ß√µes ou arquiteturas espec√≠ficas.
